```python
import os
import json
from decimal import Decimal
import pytextable
```

```python
"""
Creates the top_10 models available in the appendix of our BA
"""

exclusions = ['tex_table_top1_asset.txt','original_comparison_plot.png','original_comaprisonplot.png']


for model in os.listdir('results'):
    for asset in os.listdir(f'results/{model}'):
        for lag in os.listdir(f'results/{model}/{asset}'):
            if lag in exclusions:
                continue
            for yearcombination in os.listdir(f'results/{model}/{asset}/{lag}'):
                if yearcombination in exclusions:
                    continue
                for contentfolder in os.listdir(f'results/{model}/{asset}/{lag}/{yearcombination}'):
                    pytable_prep = []
                    top_seeds = []
                    if contentfolder == 'statistics':
                                        with open(f'results/{model}/{asset}/{lag}/{yearcombination}/{contentfolder}/mse_of_models.json','r') as file:
                                            loaded_mse_of_models = eval(json.load(file))
                                        
                                        for seed in loaded_mse_of_models:
                                            top_seeds.append((seed, loaded_mse_of_models[seed]['MSE_train_scaled']))
                                        sorted_list_with_seed = sorted(top_seeds, key = lambda x: x[1], reverse = False)[:10]
                                        for seed,_ in sorted_list_with_seed:
                                            with open(f'results/{model}/{asset}/{lag}/{yearcombination}/hitratio/{seed}.txt','r') as file:
                                                hitratios = eval(json.load(file))
                                            pytable_prep.append([seed, round(hitratios['Hit Ratio Train'],5), round(hitratios['Hit Ratio Test'],5),round(hitratios['Sharpe'],6),f"{Decimal(loaded_mse_of_models[seed]['MSE_train_scaled']):.8E}",f"{Decimal(loaded_mse_of_models[seed]['MSE_test_scaled']):.8E}"])
                                            with open(f'results/{model}/{asset}/{lag}/{yearcombination}/{contentfolder}/tex_table2.txt', 'w') as file:
                                                file.write(pytextable.tostring(
                                                pytable_prep,
                                                #fmt=".6f",
                                                header=("Seed","Hit Ratio Train", "Hit Ratio Test", "Sharpe Test", "MSE Train", "MSE Test"),
                                                caption=f"Metrics using Dataset {yearcombination} on Lag {lag}",
                                                label=f"tab:Metrics{yearcombination}_{asset}_{lag}",
                                                alignment='rrrrrr'
                                                ))
```

```python
"""
Takes all the models in a lag, in a yearcombination, and creates a tex table with the best models.
"""
exclusions = ['tex_table_top1_asset.txt','original_comparison_plot.png','original_comaprisonplot.png']


for model in os.listdir('results'):
        for asset in os.listdir(f'results/{model}'):
                for lag in os.listdir(f'results/{model}/{asset}'):
                    if lag in exclusions:
                        continue

                    top_seeds = []
                    try:
                        pytable_prep = []
                        
                        for yearcombination in os.listdir(f'results/{model}/{asset}/{lag}'):
                            top_seeds = []
                            if yearcombination in exclusions:
                                continue
                            for contentfolder in os.listdir(f'results/{model}/{asset}/{lag}/{yearcombination}'):
                                if contentfolder == 'statistics':# and asset == 'btc_hist.csv' and lag == '2' and yearcombination == '[2020]-[2021]':
                                    
                                    with open(f'results/{model}/{asset}/{lag}/{yearcombination}/{contentfolder}/mse_of_models.json','r') as file:
                                        loaded_mse_of_models = eval(json.load(file))
            
                                    for seed in loaded_mse_of_models:
                                        top_seeds.append((seed, loaded_mse_of_models[seed]['MSE_train_scaled']))
                                    sorted_list_with_seed = sorted(top_seeds, key = lambda x: x[1], reverse = False)[:1]
                                    for seed,_ in sorted_list_with_seed:
                                        with open(f'results/{model}/{asset}/{lag}/{yearcombination}/hitratio/{seed}.txt','r') as file:
                                            hitratios = eval(json.load(file))
                                        pytable_prep.append([f"{yearcombination[1:5]}-{yearcombination[-5:-1]}" , lag, round(hitratios['Hit Ratio Train'],5), round(hitratios['Hit Ratio Test'],5),round(hitratios['Sharpe'],6),f"{Decimal(loaded_mse_of_models[seed]['MSE_train_scaled']):.5E}",f"{Decimal(loaded_mse_of_models[seed]['MSE_test_scaled']):.5E}"])
    
                    except:
                        print(model, asset, lag, yearcombination)
                        pass
                    try:
                        with open(f'results/{model}/{asset}/{lag}/tex_table_top1_asset.txt', 'w') as file:
                                        file.write(pytextable.tostring(
                                            pytable_prep,
                                            #fmt=".6f",
                                            header=("Years", "Lag", "Hit Ratio Train", "Hit Ratio Test", "Sharpe Test", "MSE Train", "MSE Test"),
                                            caption=f"Metrics using Dataset {yearcombination}",
                                            label=f"tab:Metrics_{asset}",
                                            alignment='rrrrrrr'
                                            ))#Metrics using 2015, 2016 as train and 2017 as test on Lag
                    except:
                        print(model, asset, lag, yearcombination)
                        pass
```
